```julia
data=readdlm(download("https://raw.githubusercontent.com/Dinesh-Adhithya-H/Julia-Scientific-programming/main/wikipediaEVDraw.csv"),',')
```

```julia
using Dates
```

```julia
Dates.DateTime(data[1,1],"d u y")
```

```julia
data[1,1]
```

```julia
col1=data[:,1]
for i in 1:length(col1)
    col1[i]=Dates.DateTime(col1[i],"d u y")
end
```

```julia
col1
```

```julia
#Dates.datetime2rata(col1[1])
```

```julia
days(x)=Dates.datetime2rata(x)-Dates.datetime2rata(col1[54])
```

```julia
days_array=Array{Int64}(undef,54)
for i in 1:length(col1)
    days_array[i]=days(col1[i])
end
```

```julia
data[:,1]=days_array
writedlm("wikipediaEVDdatesconverted.csv", data, ',')
```
